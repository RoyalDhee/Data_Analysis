#### Vector Addition

```python
a = np.array([2, 3])
b = np.array([1, 5])
print("a + b =", a + b)
```

```python
# Vector addition
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

vector_sum = a + b
print(f"vector a: {a}")
print(f"vector b: {b}")
print(f"a + b: {vector_sum}")

# Geometric interpretation in 2D
a_2d = np.array([2, 3])
b_2d = np.array([1, 2])
sum_2d = a_2d + b_2d

plt.figure(figsize=(9, 6))
plt.quiver(0, 0, a_2d[0], a_2d[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.005)
plt.quiver(0, 0, b_2d[0], b_2d[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.005)
plt.quiver(0, 0, sum_2d[0], sum_2d[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005)
plt.quiver(a_2d[0], a_2d[1], b_2d[0], b_2d[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.005, alplha=0.5)
plt.xlim(-1, 4)
plt.ylim(-1, 6)
plt.grid(True)
plt.title("vector Addition: Parallelogram Rule")
plt.legend(['Vector a', 'Vector b', 'a + b'])
plt.show()
```

```python
# Properties of vector addition
c = np.array([1, -1, 2])

# Commutative property: a + b = b + a
print(f"a + b = {a + b}")
print(f"b + a = {b + a}")
print(f"Commutative: {np.array_equal(a + b, b + a)}")

# Associative Property: (a + b) + c = a + (b + c)
left = (a + b) + c
right = a + (b + c)

print(f"(a + b) + c = {left}")
print(f"a + (b + c) = {right}")
print(f"Associative {np.array_equal(left, right)}")
```

#### Scalar Multiplication


```python
prices = np.array([100, 200])
discount = 0.5
print("Half Price", discount * prices)
```

```python
# Scalar Multiplication
vector = np.array([2, 3, 1])
scalar = 3

scaled_vector = scalar * vector
print(f"Original vactor: {vector}")
print(f"Scalar: {scalar}")
print(f"Scaled vector: {scaled_vector}")

# Geometric interpretation
original_2d = np.array([2, 1])
scales = [0.5, 1, 2, -1]

plt.figure(figsize=(10, 6))
colors = ['red', 'blue', 'green', 'purple']
for i, scale in enumerate(scales):
    scaled = scale * original_2d
    plt.quiver(0, 0, scaled[0], scaled[1], angles='xy', scale_units='xy', scale=1, color=colors[i], width=0.005, label=f"{scale} * v")

plt.xlim(-3, 5)
plt.ylim(-3, 3)
plt.grid(True)
plt.title("Scalar Multiplication Effects")
plt.legend()
plt.show()
```

```python
# properties of scalar multiplication
a = np.array([1, 2, 3])
c1, c2 = 2, 3

# Associative: c1(c2*a) = (c1*c2)*a
left = c1 * (c2 * a)
right = (c1 * c2) * a
print(f"c1(c2*a) = {left}")
print(f"(c1*c2)*a = {right}")
print(f"Associative: {np.array_equal(left, right)}")

# Distributive over vector addition: c(a + b) = ca + cb
left = c1 * (a + b)
right = c1 * a + c1 * b
print(f"c(a + b) = {left}")
print(f"c(a + b) = {right}")
print(f"Distributive: {np.array_equal(left, right)}")
```

#### The physicist's perspective of a vector

```python
import matplotlib.pyplot as plt

# Physicist's view: vectors as arrows in space
vector_a = np.array([3, 4])
magnitude = np.linalg.norm(vector_a)
print(f"Vector: {vector_a}")
print(f"magnitude: {magnitude}")

# Direction (unit vector)
unit_vector = vector_a / magnitude
print(f"Unit vector (direction): {unit_vector}")

# Visualization
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.005)
plt.quiver(0, 0, unit_vector[0], unit_vector[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.003)
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.title("Vector: Magmitude and Direction")
plt.legend(['original Vector', 'Unit Vector'])
plt.show()
```

#### The Mathematicians perspective

```python
# Mathematician's view: vectors as abstract objects
# Example 1: Polynomials as vectors
# p(x) = 3x^2 + 2x + 1 represented as [1, 2, 3] (coefficients)
poly1 = np.array([1, 2, 3]) # 1 + 2x + 3x^2
poly2 = np.array([2, -1, 1]) # 2 - x + x^2

# Vector addition for polynomials
poly_sum = poly1 + poly2
print("Polynomial 1:", poly1)
print("Polynomial 2:", poly2)
print("Sum:", poly_sum)  # (1+2) + (2-1)x + (3+1)x^2 = 3 + x + 4x^2

# Scalar Multiplication for Polynomials
scalar = 2
poly_scaled = scalar * poly1
print(f"2 * Polynomial 1: {poly_scaled}")


# Example 2 Functiuons as vectors (fumction space)
# This is more abstract but functions can be added and scaled
def func1(x):
    return x**2

def func2(x):
    return 2*x

# Function addition: (f + g)(x) = f(x) + g(x)
def func_sum(x):
    return func1(x) + func2(x)

x_vals = np.linspace(0, 3, 100)
plt.figure(figsize=(10, 6))
plt.plot(x_vals, func1(x_vals), label='f(x) = x²')
plt.plot(x_vals, func2(x_vals), label='g(x) = 2x')
plt.plot(x_vals, func_sum(x_vals), label='(f+g)(x) = x² + 2x')
plt.legend()
plt.title("Functions as Vectors")
plt.grid(True)
plt.show()
```

#### Audio Signals

```python
# Audio signals as vectors
sample_rate = 44100
duration = 0.1  #0.1 seconds
t = np.linspace(0, duration, int(sample_rate * duration))
t

# Two sine waves (audio signals)
signal1 = 0.5 * np.sin(2 * np.pi * 440 * t)     # A note (440 Hz)
signal2 = 0.3 * np.sin(2 * np.pi * 880 * t)     # A note one octave higher

# Vector addition of audio signals
mixed_signal = signal1 + signal2

# Scalar multiplication (volume control)
quiet_signal = 0.5 * mixed_signal

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(t[:1000], signal1[:1000])
plt.title("Signal 1 (440 Hz)")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(t[:1000], signal2[:1000])
plt.title("Signal 2 (880 Hz)")
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(t[:1000], mixed_signal[:1000])
plt.title("Mixed Signal (Addition)")
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(t[:1000], quiet_signal[:1000])
plt.title("Quiet Signal (Scalar Multiplication)")
plt.grid(True)

plt.tight_layout()
plt.show()

print(f"Signal 1 shape: {signal1.shape}")
print(f"Signal 2 shape: {signal2.shape}")
print(f"Mixed signal shape: {mixed_signal.shape}")
```

#### Other Vector Examples

```python
# 1. Word frequencies in documents (Real-world)
document1 = np.array([5, 3, 0, 2, 1])   #word counts: [the, cat, dog, sat, ran]
document2 = np.array([2, 0, 4, 1, 3])   # word counts for same vocabulary

# Document similarity (dot product)
similarity = np.dot(document1, document2)
print("Document vectors:")
print("Doc 1:", document1)
print("Doc 2:", document2)
print("Similarity score:", similarity)
```

```python
# 2. Stock prices over time (Real-world)
stock_prices = np.array([100, 102, 98, 105, 103, 107])
print("\nStock price vector:", stock_prices)

# Price changes (difference vector)
price_changes = np.diff(stock_prices)
print("Price changes:", price_changes)

```

```python
# 3. Probability distributions (Abstract)
prob_dist = np.array([0.1, 0.2, 0.3, 0.25, 0.15])
print("\nProbability distribution:", prob_dist)
print("Sum (should be 1):", np.sum(prob_dist))

# Scaling maintains probability properties
scaled_prob = 2 * prob_dist
print("Scaled (no longer probability):", scaled_prob)
print("Sum:", np.sum(scaled_prob))
```

#### Matrices

```python
# Examples of notation in Numpy
A = np.array([[1, 2], [3, 4]])

print("Matrix A:")
print(A)
print("\nTranspose A^T:")
print(A.T)
print("\nInverse A^(-1):")
print(np.linalg.inv(A))
print("\nDeterminant |A|:")
print(np.linalg.det(A))
```

```python
# Creating matrices in Numpy
# Different ways to create matrices

# Method 1: From lists
matrix_a = np.array([[1, 2, 3],
                     [4, 5, 6]])

print("Matrix A (2x3):")
print(matrix_a)
print(f"Shape: {matrix_a.shape}")
```

```python
# Method 2: Using Numpy functions
zeros_matrix = np.zeros((3, 4))
ones_matrix = np.ones((2, 3))
identity_matrix = np.eye(3)
random_matrix = np.random.rand(2, 3)

print("\nZeros matrix (3x4):")
print(zeros_matrix)

print("\nOnes matrix (2x3):")
print(ones_matrix)

print("\nIdentity matrix (3x3):")
print(identity_matrix)

print("\nRandom matrix (2x3):")
print(random_matrix)
```

```python
# Matrix Terminology
print(f"\nMatrix A dimensions: {matrix_a.shape[0]} rows x {matrix_a.shape[1]} columns")
print(f"Total elements: {matrix_a.size}")

# Row and column vectors
row_vector = matrix_a[0, :]     #First row
col_vector = matrix_a[:, 0]     # First column

print(f"Row vector: {row_vector}")
print(f"Column vector: {col_vector}")
```
